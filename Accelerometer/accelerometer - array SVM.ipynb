```python
import pandas as pd
import numpy as np
all_labels = []
all_data = []
for i in range(1, 10):
    temp_data = pd.read_csv("adl-0"+str(i)+"-acc.csv", encoding='utf-8').values #stores values in array of array
    SV_total = np.array([]) #stores values of SV_total as an array
    Prev_and_next_fives = np.array([]) #stores the largest change in SV_total along with 4 previous changes.
    for j in temp_data:
        SV_total = np.append(SV_total,j[1])
    curr_max = SV_total[0]
    count = 0
    for j in range(1, len(SV_total)):
        if SV_total[j] > curr_max:
            curr_max = SV_total[j]
            count = j
    for j in range(0, 11):
        Prev_and_next_fives = np.append(Prev_and_next_fives, SV_total[count+j-5])
    #print(Prev_and_next_fives)
    all_data.append(Prev_and_next_fives)
    all_labels.append(0)
print(all_data[1])    
for i in range(10, 41):
    temp_data = pd.read_csv("adl-"+str(i)+"-acc.csv", encoding='utf-8').values
    SV_total = np.array([])
    Prev_and_next_fives = np.array([])
    for j in temp_data:
        SV_total = np.append(SV_total,j[1])
    curr_max = SV_total[0]
    count = 0
    for j in range(1, len(SV_total)):
        if SV_total[j] > curr_max:
            curr_max = SV_total[j]
            count = j
    if count+5 >= len(SV_total):
        continue
    else:
        for j in range(0, 11):
            Prev_and_next_fives = np.append(Prev_and_next_fives, SV_total[count+j-5])
        all_data.append(Prev_and_next_fives)
        all_labels.append(0)
for i in range(1, 10):
    temp_data = pd.read_csv("fall-0"+str(i)+"-acc.csv", encoding='utf-8').values
    SV_total = np.array([])
    Prev_and_next_fives = np.array([])
    for j in temp_data:
        SV_total = np.append(SV_total,j[1])
    curr_max = SV_total[0]
    count = 0
    for j in range(1, len(SV_total)):
        if SV_total[j] > curr_max:
            curr_max = SV_total[j]
            count = j
    for j in range(0, 11):
        Prev_and_next_fives = np.append(Prev_and_next_fives, SV_total[count+j-5])
    all_data.append(Prev_and_next_fives)
    all_labels.append(1)
for i in range(10, 31):
    temp_data = pd.read_csv("fall-"+str(i)+"-acc.csv", encoding='utf-8').values
    SV_total = np.array([])
    Prev_and_next_fives = np.array([])
    for j in temp_data:
        SV_total = np.append(SV_total,j[1])
    curr_max = SV_total[0]
    count = 0
    for j in range(1, len(SV_total)):
        if SV_total[j] > curr_max:
            curr_max = SV_total[j]
            count = j
    for j in range(0, 11):
        Prev_and_next_fives = np.append(Prev_and_next_fives, SV_total[count+j-5])
    all_data.append(Prev_and_next_fives)
    all_labels.append(1)        
all_data = np.array(all_data)
```

```python
all_SVs = all_data
all_labels = np.array(all_labels)
print(all_SVs)
```

```python
from sklearn.utils import shuffle
all_SVs = shuffle(all_SVs, random_state = 0)
all_labels = shuffle(all_labels, random_state = 0)
print(all_SVs[14])
print(all_labels)
```

```python
print (all_SVs.shape)
print (all_labels.shape)
```

```python
train_x = all_SVs[:40]
train_y = all_labels[:40]
test_x = all_SVs[40:]
test_y = all_labels[40:]
print ("# of training data is %d" %len(train_x))
print ("# of testing data is %d" %len(test_x))
```

```python
from sklearn import svm
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
Kernel = {"linear", "poly", "rbf", "sigmoid", "precomputed"}
C = range(1,21)
Degree = {1, 2, 3, 4, 5, 6}
Gamma = {"auto", "scale", "auto_deprecated"}
Coef0 = {0}
Shrinking = {True, False}
Probability = {True, False}
Verbose = {True, False}
max_acc = 0
k_tracker = ""
C_tracker = 0
d_tracker = 0
g_tracker = ""
co_tracker = 0
s_tracker = True
p_tracker = False
v_tracker = True
for k in Kernel:
    for c in C:
        for d in Degree:
            for g in Gamma:
                for co in Coef0:
                    for s in Shrinking:
                        for p in Probability:
                            for v in Verbose:
                                try:
                                    lin_clf = svm.SVC(C = c/10, kernel = k, degree = d, gamma = g, coef0 = co/10,shrinking = s,probability = p,verbose = v)
                                    # fit( ) will do the model training, i.e., learn the model parameters
                                    lin_clf.fit(train_x, train_y)
                                    from sklearn.metrics import accuracy_score
                                    # predict( ) will do the model prediction, predict y based on the input x
                                    predict_y = lin_clf.predict(test_x)
                                    if accuracy_score(predict_y, test_y)>=max_acc:
                                        max_acc = accuracy_score(predict_y,test_y)
                                        C_tracker = c/10
                                        k_tracker = k
                                        d_tracker = d
                                        g_tracker = g
                                        co_tracker = co
                                        s_tracker = s
                                        p_tracker = p
                                        v_tracker = v
                                    print(accuracy_score(predict_y,test_y))
                                except:
                                    pass
print ('testing acc is %f' %max_acc)
print(C_tracker)
print(k_tracker)
print(d_tracker)
print(g_tracker)
print(co_tracker)
print(s_tracker)
print(p_tracker)
print(v_tracker)
```

```python
from sklearn.metrics import accuracy_score, log_loss
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC, NuSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis

classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="rbf", C=0.025, probability=True),
    NuSVC(probability=True),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    GaussianNB(),
    LinearDiscriminantAnalysis(),
    QuadraticDiscriminantAnalysis()]

# Logging for Visual Comparison
log_cols=["Classifier", "Accuracy", "Log Loss"]
log = pd.DataFrame(columns=log_cols)

for clf in classifiers:
    clf.fit(train_x, train_y)
    name = clf.__class__.__name__
    
    print("="*30)
    print(name)
    
    print('****Results****')
    train_predictions = clf.predict(test_x)
    acc = accuracy_score(test_y, train_predictions)
    print("Accuracy: {:.4%}".format(acc))
    
    train_predictions = clf.predict_proba(test_x)
    ll = log_loss(test_y, train_predictions)
    print("Log Loss: {}".format(ll))
    
    log_entry = pd.DataFrame([[name, acc*100, ll]], columns=log_cols)
    log = log.append(log_entry)
    
print("="*30)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_color_codes("muted")
sns.barplot(x='Accuracy', y='Classifier', data=log, color="b")

plt.xlabel('Accuracy %')
plt.title('Classifier Accuracy')
plt.show()

sns.set_color_codes("muted")
sns.barplot(x='Log Loss', y='Classifier', data=log, color="g")

plt.xlabel('Log Loss')
plt.title('Classifier Log Loss')
plt.show()
```
